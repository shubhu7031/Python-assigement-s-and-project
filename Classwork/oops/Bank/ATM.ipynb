```python
#ATM System using python

c=1
pin = 3214
balance=5000

while c<=3:
  pina=int(input("\nEnter pin:- "))
  if pina == pin:
    print("\nWelcome...!\n")
    print("Choose Transaction:-")
    choice=""
    while choice!='quit':
      choice=input(("1. Check Balance.\n""2. Deposit.\n""3. Withdraw.\n""\nEnter 'quit' to exit."))
      if choice==str(1):
        print("\nYour balance is,",balance,"\n")
      elif choice == str(2):
        depo=int(input("\nEnter amount to deposit:- "))
        balance=depo+balance
        print("\nYour Current Balance is,",balance,"\n")
      elif choice==str(3):
        withd=int(input("\nEnter amount to Withdraw:- "))
        if withd>balance:
          print("\nInsufficient Balance, please enter amt. as your balance.!")
        else:
          balance=balance-withd
        print("\nYour Current balance is,",balance,"\n")
      else:
        print("Please Select a Valid Option...!")
    break
    
  else:
    print("\nWrong pin...!")
    if c==3:
      print("max attempts Done...!")
      print("Try after 24 hrs.")
  c+=1
```

#ATM System Using Class & Methods.

```python
class atm:
  c=0
  pin=3214
  balance=5000
  def greet(self):
    print("Welcome to MyBank ATM...!\n")
  def closing(self):
    print("\nThanks for using ATM...!")

  def option(self):
    opt=int(input("Choose Transaction:-\n1. Check Balance.\n""2. Deposit.\n""3. Withdraw.\n""4. Exit.\nChoose Option :- "))

    if opt==1:
      return atm.check_bal(self)
    elif opt==2:
      return atm.deposit(self)
    elif opt==3:
      return atm.withdraw(self)
    elif opt==4:
      return atm.closing(self)
    else:
      print("Invalid Input, please Select a Valid Input..!\n")
      return atm.option(self)
    

  def login(self):
    
    pina=int(input("Enter Your Pin:- "))
    if pina==self.pin:
      print("Welcome...!\n")
      return atm.option(self)
    else:
      self.c+=1
      if self.c==3:
        print("\nMax attempts Done...!\nTry After 24 HRs.\n")
        atm.closing(self)
      else:
        print("Wrong Pin..!\n")
        return atm.login(self)
  
  def check_bal(self):
    print("\nYour Current Balance is,",self.balance)
    return atm.option(self)
  
  def deposit(self):
    depo=int(input("\nEnter Amount to Deposit:- "))
    self.balance+=depo
    print(f"\nYour Current Balance is {self.balance}.\n")
    return atm.option(self)

  def withdraw(self):
    withd=int(input("\nEnter Amount to withraw:- "))
    if withd>self.balance:
      print("Insufficient Balance.\n")
      return atm.option(self)
    else:
      self.balance-=withd
      print("Your Current Balance is,",self.balance,"\n")
      return atm.option(self)

customer=atm()
customer.login()
```
